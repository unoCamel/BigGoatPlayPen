```python
# IMPORTS
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# DATA
df = pd.read_csv("datasets/insurance.csv")
```

```python
dfMales = df[df['sex']=='male']
dfFemales = df[df['sex']=='female']
```

```python
dfFemales.head()
df['age'].count()
#df['age'].max()
```

```python
#%matplotlib notebook
#Average plots across ages

  

plt.scatter(df['age'], df['charges'])
plt.ylabel('Charges')
plt.show()

```

```python
meanCharge = {x: df.loc[df['age'] == x].mean() for x in range(18,65)}  
plt.plot(meanCharge.keys(), meanCharge.values())
plt.ylabel('Charges')
plt.show()
```

```python
ages = list(df.age.unique())
ages.sort()
means = [df[df.age==a].charges.mean() for a in ages]
plt.plot(ages,means,label='mean')
plt.scatter(df.age.values,df.charges.values,alpha=0.3,color='pink',label='raw data')
plt.xlabel("age")
plt.ylabel("mean")
plt.title("age vs mean")
plt.legend()
plt.show()
```
